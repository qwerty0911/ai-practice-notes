프로젝트 분량보다 **도메인 선택과 프롬프트 다듬기**, **시연 / 발표** 에 시간을 더 쓰는 구성입니다.

발표는 **1인당 3~5분**, 흐름은 슬라이드 5장 그대로 따라가면 됩니다.

## 4. 도메인 선택 가이드

**좋은 도메인의 조건**

- 답변 톤이 일반 챗봇과 분명히 달라야 함 (System Prompt 의 효과가 드러나야 함)
- 데모 질문 5개를 떠올리기 쉬운 주제
- 모델이 그 분야의 기본 지식을 가지고 있을 것 (너무 좁은 사내 도메인 X)

**예시 4개 (그대로 골라도 좋고, 변형해도 좋습니다)**

1. **법률 상담 챗** — 일반인 대상으로 법률 용어를 쉽게 풀어 설명
2. **요리 레시피 도우미** — 재료 입력 → 만들 수 있는 요리 제안
3. **여행지 추천** — 조건 입력 → 국내 여행지 추천
4. **코드 리뷰 봇** — 짧은 코드 스니펫 → 개선점 한 줄

본인이 평소 관심 있던 주제(취미·전공·직무) 가 있으면 그쪽이 더 좋습니다. 데모 질문을 만들기 훨씬 쉽거든요.

## 5. System Prompt 작성 팁

5-3 에서 배운 권장 순서를 그대로 적용합니다.

**순서: 역할 → 톤 → 언어 → 제약**

- **역할**: "너는 OO 분야의 OO 야"
- **톤**: "친절하게 / 간결하게 / 단호하게"
- **언어**: "항상 한국어로 답해"
- **제약**: "3문장 이내 / 코드 예시 없이 / 200자 이내" 등

**예시 (법률 상담 챗)**

```python
SYSTEM_PROMPT = (
    "너는 일반인에게 법률 용어를 쉽게 풀어 설명하는 상담가야. "
    "전문 용어가 나오면 한 줄 풀이를 곁들이고, "
    "답변은 한국어로 친절하게, 한 번에 3문장 이내로 끝내. "
    "법적 판단이 필요한 사안에는 반드시 '실제 사안은 전문가 상담을 권합니다' 라고 덧붙여."
)
```

**자주 하는 실수**

- 너무 김 — 모델이 일부만 따라갑니다. 핵심 4가지(역할/톤/언어/제약) 위주로.
- 부정문 위주 — "X 하지 마" 보다 "Y 만 해" 가 더 잘 먹습니다.
- 도메인이 안 드러남 — "친절한 어시스턴트야" 만으로는 일반 챗봇과 구분이 안 됩니다.

## 6. 데모 질문 5개 작성 팁

발표 때 시연할 질문 5개입니다. 아래 구성을 따르면 챗봇을 골고루 보여줄 수 있습니다.

- **#1 쉬운 질문** — 챗봇이 가장 자신 있게 답할 수 있는 것. 첫 시연이 매끄럽게 시작됩니다.
- **#2~#3 도메인 핵심 질문** — 이 챗봇의 존재 이유를 보여주는 질문. 일반 챗봇과 답이 분명히 달라야 합니다.
- **#4 경계 케이스** — 약간 까다롭거나 모호한 질문. 챗봇이 어디까지 버티는지 드러납니다.
- **#5 영역 밖 질문** — 도메인을 벗어난 질문. System Prompt 의 제약(거절·다른 주제 안내)이 잘 작동하는지 확인합니다.

질문은 짧게, 출력이 발표 화면에 한눈에 들어오게 고르세요.

**예시 (법률 상담 챗)**

1. (쉬움) "전세 계약할 때 가장 먼저 확인해야 할 게 뭐야?"
2. (핵심) "근저당권이랑 저당권 차이를 한 문장으로 설명해줘"
3. (핵심) "임대인이 보증금을 안 돌려주면 어떻게 해야 해?"
4. (경계) "내가 직접 쓴 차용증도 법적 효력이 있어?"
5. (영역 밖) "오늘 점심 메뉴 좀 추천해줘"

## 10. 잘 안 풀릴 때

- `cd mini_project` 후에 `uvicorn app.main:app` 명령을 실행했는지 확인
- 포트 충돌이면 시작 코드 노트북의 1단계 `PORT` 만 바꾸고 config / 서버 띄우기 셀 재실행

**`/chat/stream` 호출 시 `[ERROR]` 가 나옴**

- 본문 끝의 `[ERROR] {타입}: {메시지}` 를 그대로 읽으면 됩니다. 대부분 API_KEY / MODEL / BASE_URL 설정 문제입니다.
- 시작 코드 노트북 1단계 셀로 돌아가서 값 확인 후 config 셀 재실행
