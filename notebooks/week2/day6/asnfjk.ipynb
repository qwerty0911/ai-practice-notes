```python
import numpy as np
import scipy as sp

class Perceptron():
    def __init__(self, X, Y, Lr = 0.1, iteration = 500):
        self.x = X
        self.y = Y
        self.lr = Lr
        self.iteration = iteration

    def Model(self):
        return np.matmul(self.x, self.weight) + self.bias

    def affine(self):
        pass

    def activate(self, X):
        return sp.special.expit(X)

    def forward(self):
        output = self.Model()
        return self.activate(output)

    def loss(self, Ytgt, Ypredf): #categorical cross entropy
        return -np.mean(Ytgt * np.log(Ypredf) + (1 - Ytgt) * np.log(1 - Ypredf))

    def run(self):
        loss_score = []
        for i in range(self.iteration):
            self.chache = []

            Ypred = self.forward()
            loss = self.loss(self.y, Ypred)
            loss_score.append(loss)
            print(f"iteration: {i}, loss: {loss}")
            dZ = Ypred - self.y
            X = np.array(self.x)
            dW = np.dot(X.T, dZ) / len(self.x)
            db = np.sum(dZ) / len(self.x)
            self.weight -= self.lr * dW
            self.bias -= self.lr * db
        return loss_score
```
